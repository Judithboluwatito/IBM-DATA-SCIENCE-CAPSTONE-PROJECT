```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
launch_site_lat = 28.562
launch_site_lon = -80.577

coastline_lat = 28.56247
coastline_lon = -80.5707

distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
print("Distance between launch site and closest coastline point is: {:.2f} km".format(distance_coastline))
```

HIGHWAY

```python
import folium
from folium.plugins import MousePosition
from folium.features import DivIcon

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Define the launch site and highway coordinates
launch_site_lat = 28.57336
launch_site_lon = -80.65547

highway_lat = 28.5248
highway_lon = -80.6774

# Create a folium Map object centered on the launch site coordinates
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=10)

# Add a marker for the launch site and highway coordinates
folium.Marker(
    location=[launch_site_lat, launch_site_lon],
    icon=folium.Icon(color='green'),
    popup='KSC LC-39A'
).add_to(site_map)

folium.Marker(
    location=[highway_lat, highway_lon],
    icon=folium.Icon(color='red'),
    popup='Highway'
).add_to(site_map)

# Calculate the distance between the launch site and highway coordinates
distance_highway = calculate_distance(launch_site_lat, launch_site_lon, highway_lat, highway_lon)

# Create a marker for the distance between the launch site and highway coordinates
distance_marker = folium.Marker(
    location=[(launch_site_lat + highway_lat)/2, (launch_site_lon + highway_lon)/2],
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_highway),
        )
    )
distance_marker.add_to(site_map)

# Create a PolyLine between the launch site and highway coordinates
lines=folium.PolyLine(locations=[[launch_site_lat, launch_site_lon],[highway_lat, highway_lon]], weight=5, color='blue')
site_map.add_child(lines)

# Add a MousePosition plugin to display the latitude and longitude of the cursor position
MousePosition().add_to(site_map)

# Display the map
site_map
```
